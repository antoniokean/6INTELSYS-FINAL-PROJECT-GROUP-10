# 01 EDA

This notebook provides a lightweight exploratory data analysis pass for the Speech Commands project. It summarizes the manifest, checks split and label coverage, and visualizes a few example log-Mel spectrograms.

```python
import csv
from collections import Counter
from pathlib import Path

import librosa
import matplotlib.pyplot as plt
import numpy as np

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / "data").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_DIR = PROJECT_ROOT / "data"
MANIFEST_PATH = DATA_DIR / "dataset_manifest.csv"
MANIFEST_PATH
```

```python
with MANIFEST_PATH.open("r", encoding="utf-8", newline="") as file:
    rows = list(csv.DictReader(file))

keep_rows = [row for row in rows if row["status"] == "keep"]
split_counts = Counter(row["split"] for row in rows)
status_counts = Counter(row["status"] for row in rows)
label_counts = Counter(row["label"] for row in keep_rows)

print(f"Total samples: {len(rows)}")
print(
    "Split counts: "
    + ", ".join(f"{split}={count}" for split, count in split_counts.items())
)
print(
    "Status counts: "
    + ", ".join(f"{status}={count}" for status, count in status_counts.items())
)
print(f"Unique labels: {len(label_counts)}")
print("Top 10 labels by kept samples:")
for label, count in label_counts.most_common(10):
    print(f" - {label}: {count}")
```

```python
figure, axes = plt.subplots(1, 2, figsize=(14, 4))

split_names = ["training", "validation", "testing"]
axes[0].bar(split_names, [split_counts.get(name, 0) for name in split_names], color=["#3b82f6", "#10b981", "#f59e0b"])
axes[0].set_title("Samples per Split")
axes[0].set_ylabel("Count")
axes[0].grid(axis="y", alpha=0.3)

top_labels = label_counts.most_common(12)
axes[1].bar([label for label, _ in top_labels], [count for _, count in top_labels], color="#8b5cf6")
axes[1].set_title("Top 12 Labels by Kept Samples")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis="x", rotation=45)
axes[1].grid(axis="y", alpha=0.3)

figure.tight_layout()
plt.show()
```

```python
SAMPLE_RATE = 16000
N_MELS = 64
N_FFT = 1024
HOP_LENGTH = 256
TARGET_NUM_SAMPLES = SAMPLE_RATE


def load_audio(path):
    audio, _ = librosa.load(path, sr=SAMPLE_RATE, mono=True)
    if len(audio) < TARGET_NUM_SAMPLES:
        audio = np.pad(audio, (0, TARGET_NUM_SAMPLES - len(audio)))
    else:
        audio = audio[:TARGET_NUM_SAMPLES]
    return audio


def audio_to_log_mel(audio):
    mel = librosa.feature.melspectrogram(
        y=audio,
        sr=SAMPLE_RATE,
        n_mels=N_MELS,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
    )
    return librosa.power_to_db(mel, ref=np.max)


example_rows = []
seen_labels = set()
for row in keep_rows:
    if row["split"] != "training":
        continue
    if row["label"] in seen_labels:
        continue
    example_rows.append(row)
    seen_labels.add(row["label"])
    if len(example_rows) == 4:
        break

figure, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for axis, row in zip(axes, example_rows):
    features = audio_to_log_mel(load_audio(DATA_DIR / row["path"]))
    image = axis.imshow(features, aspect="auto", origin="lower", cmap="magma")
    axis.set_title(f"{row['label']} | {Path(row['path']).name}")
    axis.set_xlabel("Time")
    axis.set_ylabel("Mel Bin")

for axis in axes[len(example_rows):]:
    axis.axis("off")

figure.colorbar(image, ax=axes.tolist(), fraction=0.03, pad=0.02, label="dB")
figure.suptitle("Example Log-Mel Spectrograms", fontsize=14)
figure.tight_layout(rect=(0, 0, 0.95, 0.95))
plt.show()
```
